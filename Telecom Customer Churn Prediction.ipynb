```python
tchurn['TotalCharges'] = tchurn["TotalCharges"].replace(" ",np.nan)
```

```python
tchurn = tchurn[tchurn["TotalCharges"].notnull()]
tchurn = tchurn.reset_index()[tchurn.columns]

tchurn["TotalCharges"] = tchurn["TotalCharges"].astype(float)
```

```python
tchurn.isna().sum()
```

```python
tchurn["SeniorCitizen"] = tchurn["SeniorCitizen"].replace({1:"Yes",0:"No"})
```

```python
col_to_replace = ['StreamingMovies','OnlineSecurity','OnlineBackup','DeviceProtection','TechSupport',
                  'StreamingTV']
for i in col_to_replace:
    tchurn[i]=tchurn[i].replace({"No internet service":"No"})
```

```python
tchurn["tenure"].max()
```

```python
def tenure_range(tchurn):
    if tchurn["tenure"] <= 12:
        return("0-1 Year")
    
    elif tchurn["tenure"] >12 and tchurn["tenure"] <=24:
        return("1-2 Year")
    
    elif tchurn["tenure"] >24 and tchurn["tenure"] <=36:
        return("2-3 Year")
    
    elif tchurn["tenure"] >36 and tchurn["tenure"] <=48:
        return("3-4 Year")
    
    elif tchurn["tenure"] > 48 and tchurn["tenure"] <=60:
        return("4-5 Year")
    
    elif tchurn["tenure"] >60 and tchurn["tenure"] <=72:
        return("5-6 Year")
```

```python
tchurn["tenure"]=tchurn.apply(lambda tchurn:tenure_range(tchurn),axis=1)
```
